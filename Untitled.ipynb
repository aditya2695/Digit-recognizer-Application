```python


def load_mnist(path, kind='train'):
    """Load MNIST data from `path`"""
    labels_path = os.path.join(path, 
                               '%s-labels-idx1-ubyte' % kind)
    images_path = os.path.join(path, 
                               '%s-images-idx3-ubyte' % kind)
        
    with open(labels_path, 'rb') as lbpath:
        magic, n = struct.unpack('>II', 
                                 lbpath.read(8))
        labels = np.fromfile(lbpath, 
                             dtype=np.uint8)

    with open(images_path, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII", 
                                               imgpath.read(16))
        images = np.fromfile(imgpath, 
                             dtype=np.uint8).reshape(len(labels), 784)
        images = ((images / 255.) - .5) * 2
 
    return images, labels



```

```python
# unzips mnist

import sys
import gzip
import shutil

if (sys.version_info > (3, 0)):
    writemode = 'wb'
else:
    writemode = 'w'


zipped_mnist = [f for f in os.listdir('./data/') if f.endswith('ubyte.gz')]

zipped_mnist
```

```python
def extract_files(files):
    
    for z in zipped_mnist:
        z = 'data/'+z
        with gzip.GzipFile(z, mode='rb') as decompressed, open(z[:-3], writemode) as outfile:
            outfile.write(decompressed.read()) 
            z=''
            
extract_files(zipped_mnist)
```

```python

X_train, y_train = load_mnist('data/', kind='train')
print('Rows: %d, columns: %d' % (X_train.shape[0], X_train.shape[1]))


### Mnist Data as 10 rows of 5 columns 

import matplotlib.pyplot as plt



fig, ax = plt.subplots(nrows=10, ncols=5, sharex=True, sharey=True,figsize=(15,30))

# iterating over the each class i.e. the rows
for i in range(10):
    # iterating over the first five data points i.e. the columns
    for j in range(5):
        img = X_train[y_train == i][j].reshape(28, 28)
        ax[i,j].imshow(img, cmap='Greys')
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])



plt.tight_layout()
plt.savefig('mnist_output.pdf', dpi=300) # saving output image as pdf
plt.show()
```
